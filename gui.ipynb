```python
import pandas as pd
from sklearn import linear_model
import statsmodels.api as sm
import tkinter as tk 
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
```

```python
hour= pd.read_csv('hour.csv')
day=pd.read_csv('day.csv')
```

```python
X= hour.drop(['instant','cnt','dteday','holiday','weekday','workingday','windspeed','registered','casual'],axis=1)
Y= hour['cnt']

# with sklearn
regr = linear_model.LinearRegression()
regr.fit(X, Y)


# tkinter GUI
root= tk.Tk()

canvas1 = tk.Canvas(root, width = 500, height = 350)
canvas1.pack()


# New_season label and input box
label1 = tk.Label(root, text='Season:')
canvas1.create_window(100, 100, window=label1)

entry1 = tk.Entry (root) # create 1st entry box
canvas1.create_window(270, 100, window=entry1)

# New_yr label and input box
label2 = tk.Label(root, text=' Year:')
canvas1.create_window(100, 120, window=label2)

entry2 = tk.Entry (root) # create 2nd entry box
canvas1.create_window(270, 120, window=entry2)


# New_mnth label and input box
label3 = tk.Label(root, text='Month:')
canvas1.create_window(100, 140, window=label3)

entry3 = tk.Entry (root) # create 1st entry box
canvas1.create_window(270, 140, window=entry3)

# New_hr label and input box
label4 = tk.Label(root, text='Hour:')
canvas1.create_window(100, 160, window=label4)

entry4 = tk.Entry (root) # create 2nd entry box
canvas1.create_window(270, 160, window=entry4)


# New_weathersit label and input box
label5 = tk.Label(root, text='Weather:')
canvas1.create_window(100, 180, window=label5)

entry5 = tk.Entry (root) # create 1st entry box
canvas1.create_window(270, 180, window=entry5)

# New_temp label and input box
label6 = tk.Label(root, text=' Temperature in Celsius:')
canvas1.create_window(100, 200, window=label6)

entry6 = tk.Entry (root) # create 2nd entry box
canvas1.create_window(270, 200, window=entry6)

# New_atemp label and input box
label7 = tk.Label(root, text='Natural Feel in Celsius:')
canvas1.create_window(100, 220, window=label7)

entry7 = tk.Entry (root) # create 1st entry box
canvas1.create_window(270, 220, window=entry7)

# New_hum label and input box
label8 = tk.Label(root, text=' Type Humidity:')
canvas1.create_window(100, 240, window=label8)

entry8 = tk.Entry (root) # create 2nd entry box
canvas1.create_window(270, 240, window=entry8)


def values(): 
    global New_seaon #our 1st input variable
    New_seaon = float(entry1.get()) 
    
    global New_yr #our 2nd input variable
    New_yr = float(entry2.get()) 
    
    global New_mnth #our 3rd input variable
    New_mnth = float(entry3.get()) 
    
    global New_hr #our 4th input variable
    New_hr = float(entry4.get()) 
    
    global weathersit #our 5th input variable
    New_weathersit = float(entry5.get()) 
    
    global New_temp #our 6th input variable
    New_temp = float(entry6.get())/41 
    
    global New_atemp #our 7th input variable
    New_atemp = float(entry7.get())/50 
    
    global New_hum #our 8th input variable
    New_hum = float(entry8.get())/100 
    
    
    Prediction_result  = ('Predicted Utilization per Hour ', regr.predict([[New_seaon ,New_yr,New_mnth,New_hr,
     New_weathersit, New_temp, New_atemp,New_hum]]))
                                                               
    label_Prediction = tk.Label(root, text= Prediction_result, bg='orange')
    canvas1.create_window(270, 320, window=label_Prediction)
    
button1 = tk.Button (root, text='Predict Hourly Utilization',command=values, bg='orange') # button to call the 'values' command above 
canvas1.create_window(270, 290, window=button1)
 

#plot 1st scatter 
figure3 = plt.Figure(figsize=(5,4), dpi=100)
ax3 = figure3.add_subplot(111)
ax3.scatter(day['cnt'].astype(float),day['temp'].astype(float), color = 'r')
scatter3 = FigureCanvasTkAgg(figure3, root) 
scatter3.get_tk_widget().pack(side=tk.RIGHT, fill=tk.BOTH)
ax3.legend() 
ax3.set_xlabel('Daily Utilization')
ax3.set_title('Daily Utilization Vs. Temperature')

#plot 2nd scatter 
figure4 = plt.Figure(figsize=(5,4), dpi=100)
ax4 = figure4.add_subplot(111)
ax4.scatter(hour['cnt'].astype(float),hour['hr'].astype(float), color = 'g')
scatter4 = FigureCanvasTkAgg(figure4, root) 
scatter4.get_tk_widget().pack(side=tk.RIGHT, fill=tk.BOTH)
ax4.legend() 
ax4.set_xlabel('Year')
ax4.set_title('Hourly Utilization Vs. Hour')

root.mainloop()
```
